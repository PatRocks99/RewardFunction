```python
import h5py
import numpy as np  

def find_observations(h5_path, name_hints=("observations","obs","data","X")):
    hits = []
    with h5py.File(h5_path, "r") as f:
        def visit(name, obj):
            if isinstance(obj, h5py.Dataset):
                base = name.rsplit("/", 1)[-1].lower()
                is_numeric = np.issubdtype(obj.dtype, np.number)
                looks_tabular = obj.ndim in (1, 2)
                hinted = any(h in base for h in name_hints) or \
                         str(obj.attrs.get("role","")).lower() == "observations"
                if (hinted or looks_tabular) and is_numeric:
                    hits.append(("/" + name, obj.shape, str(obj.dtype)))
        f.visititems(visit)
    return hits

print(find_observations("Training/full_data/ftg_1_0.h5"))
```
