Check the lenght of State and Action Pairs

```python
import os
import pickle

def inspect_pkl_file(file_path):
    print(f"\n--- Inspecting: {file_path} ---")
    try:
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
            print("Data type:", type(data))

            if isinstance(data, dict):
                print("Keys:", list(data.keys()))
                for key, value in data.items():
                    try:
                        length = len(value)
                    except TypeError:
                        length = "N/A (not sized)"
                    print(f"  Key: '{key}' -> Length: {length}")
            elif isinstance(data, list):
                print("Length:", len(data))
                print("First item:", data[0] if data else "Empty list")
            else:
                print("Content:", data)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

def inspect_all_pkl_files(folder_path):
    if not os.path.exists(folder_path):
        print(f"Folder '{folder_path}' does not exist.")
        return

    files = [f for f in os.listdir(folder_path) if f.endswith(".pkl")]
    if not files:
        print(f"No .pkl files found in '{folder_path}'.")
        return

    for file_name in sorted(files):
        full_path = os.path.join(folder_path, file_name)
        inspect_pkl_file(full_path)

# Example usage:
inspect_all_pkl_files('clean_data')
```
